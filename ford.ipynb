5) Create count plots for all categorical columns (fuelType, transmission, model, etc.)
using seaborn.
Analyze which categories are most common.Write insights about market trends based on these plots in comments.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("ford car.csv")

# Remove duplicate rows
df = df.drop_duplicates().reset_index(drop=True)

# Plot style
sns.set_style("whitegrid")

# Create figure
fig, axes = plt.subplots(3, 1, figsize=(12, 16))

# Fuel Type
sns.countplot(
    data=df,
    x="fuelType",
    order=df["fuelType"].value_counts().index,
    ax=axes[0]
)
axes[0].set_title("Cars by Fuel Type")

# Transmission
sns.countplot(
    data=df,
    x="transmission",
    order=df["transmission"].value_counts().index,
    ax=axes[1]
)
axes[1].set_title("Cars by Transmission")

# Model
sns.countplot(
    data=df,
    y="model",
    order=df["model"].value_counts().index,
    ax=axes[2]
)
axes[2].set_title("Cars by Model")

plt.tight_layout()
plt.savefig("Q5_countplots.png", dpi=120)
plt.show()

# Display counts
print("Fuel Type Counts:")
print(df["fuelType"].value_counts())

print("Transmission Counts:")
print(df["transmission"].value_counts())

print("Top 5 Models:")
print(df["model"].value_counts().head())

# Most common: Petrol, Manual, and Fiesta/Focus are the top categories
# in their respective columns - reflecting a mainstream, budget-to-mid
# range used-car market rather than a premium or electrified one.
```

6) Create a heatmap of the correlation matrix for numeric features.
Use seaborn with proper annotations.
Identify which features are highly correlated with the target variable (price).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load the dataset
df = pd.read_csv('ford car.csv')


# remove duplicate rows
df.drop_duplicates(inplace=True)

# select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# correlation matrix
corr_matrix = df[numeric_cols].corr()

# plot style 
sns.set(style="whitegrid")

# heatmap
plt.figure(figsize=(9 , 7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Features')
plt.tight_layout()
plt.savefig("correlation_heatmap.png")
plt.show()

# year has the strongest positive correlation with price, mileage has
# the strongest negative correlation - these two are the biggest
# drivers of price in this dataset.
```

8) Convert all categorical columns into numeric form using appropriate encoding
techniques:
Use One Hot Encoding.
Show before and after transformation for at least 2 columns.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('ford car.csv')

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Before encoding
print("Before Encoding - fuelType:")
print(df[['fuelType']].head())

print()

print("Before Encoding - transmission:")
print(df[['transmission']].head())

# One-Hot Encoding for all categorical columns
df_encoded = pd.get_dummies(df, drop_first=True)

# After encoding
fuel_cols = [col for col in df_encoded.columns if col.startswith("fuelType_")]
trans_cols = [col for col in df_encoded.columns if col.startswith("transmission_")]

print()

print("After Encoding - fuelType:")
print(df_encoded[fuel_cols].head())

print()

print("After Encoding - transmission:")
print(df_encoded[trans_cols].head())

print()

print("Shape before encoding:", df.shape)
print("Shape after encoding:", df_encoded.shape)


# Used One Hot Encoding since model, transmission, and fuelType are
# nominal categories (no natural order). drop_first=True drops one
# category per column to avoid the dummy variable trap.



```

9) Apply Standard Scaling (using StandardScaler from sklearn) on the numeric
independent features.

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler

# load the dataset 
df = pd.read_csv('ford car.csv')

# remove duplicate rows 
df.drop_duplicates(inplace=True)

# numeric indipendent features 
numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# apply standard scaling 
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numeric_features])

# convert the scaled data back to a DataFrame 
scaled_df = pd.DataFrame(scaled_data, columns=numeric_features)
print("first 5 rows of scaled data")
print(scaled_df.head())




```

. Load and clean the data (handle missing & duplicate values).
. Perform EDA (histograms + count plots + heatmap).
. Identify input and output features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("ford car.csv")

print("Missing values:")
print(df.isnull().sum())

print()

print("Duplicate rows:", df.duplicated().sum())

df = df.drop_duplicates().reset_index(drop=True)

print("Shape after cleaning:", df.shape)

sns.set_style("whitegrid")

# Histograms
num_cols = ["price", "mileage", "year", "engineSize", "mpg"]

fig, axes = plt.subplots(2, 3, figsize=(16, 9))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col], bins=40)
    axes[i].set_title(col)

fig.delaxes(axes[5])
plt.tight_layout()
plt.savefig("histograms.png")
plt.show()

# Count plots
fig, axes = plt.subplots(3, 1, figsize=(12, 16))

sns.countplot(data=df, x="fuelType", order=df["fuelType"].value_counts().index, ax=axes[0])
axes[0].set_title("Fuel Type")

sns.countplot(data=df, x="transmission", order=df["transmission"].value_counts().index, ax=axes[1])
axes[1].set_title("Transmission")

sns.countplot(data=df, y="model", order=df["model"].value_counts().index, ax=axes[2])
axes[2].set_title("Model")

plt.tight_layout()
plt.savefig("countplots.png")
plt.show()

# Correlation heatmap
corr = df.select_dtypes(include="number").corr()

plt.figure(figsize=(9, 7))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("heatmap.png")
plt.show()

# Features and target
target = "price"
features = df.columns.drop(target)

print("Independent Features:")
print(list(features))

print()

print("Dependent Feature:")
print(target)

# One-Hot Encoding
df_encoded = pd.get_dummies(
    df,
    columns=["model", "transmission", "fuelType"],
    drop_first=True
)

print()

print("Shape after encoding:", df_encoded.shape)

# Feature Scaling
scaler = StandardScaler()

numeric_cols = ["year", "mileage", "tax", "mpg", "engineSize"]

df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])

print()

print("Scaled numeric features:")
print(df_encoded[numeric_cols].head())

# Final data
X = df_encoded.drop("price", axis=1)
y = df_encoded["price"]

print()

print("X Shape:", X.shape)
print("y Shape:", y.shape)
```
